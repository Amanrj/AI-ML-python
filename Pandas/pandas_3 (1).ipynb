```python
backfill in pandas involes filling missing values in a df or series using next valid observation to fill the gap 
```

```python
data = {
    'a' : [1, None, 3, None,5],
    'b' : [None, 2, None,4, None]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.fillna(method = 'bfill')
```

```python
titanic_train.head()
```

```python
# checking uniqe value of any column
```

```python
titanic_train['Pclass'].unique()
```

```python

new_survived=pd.Categorical(titanic_train['Survived'])
new_survived=new_survived.rename_categories(['Died','Survived'])
new_survived.describe()
```
